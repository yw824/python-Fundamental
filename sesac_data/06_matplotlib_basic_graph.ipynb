### 3.5 세 가지 그래프를 한 번에 그리고 옵션 추가하기
```
plt.title("Test Graph", size=15)                 # 제목
plt.xlabel('x')                                  # x축 범례
plt.ylabel('y')                                  # y축 범례

````python
plt.scatter(x, y)
plt.bar(x, y)
plt.plot(x, y)
plt.show()
````

````python
plt.scatter(x, y)
plt.bar(x, y)
plt.plot(x, y)

plt.title("Test Graph", size=15)                 # 제목
plt.xlabel('x')                                  # x축 범례
plt.ylabel('y')                                  # y축 범례

plt.show()
````

각 그래프에 대해서도 범례를 달 수 있다.
- 각 그래프에 대한 범례를 달기만 하고 보여주지 않으면 보이지 않는다. 
- 이를 위해 아래의 코드를 한 줄 추가해야 한다. <br>
``` plt.legend() ```

````python
plt.scatter(x, y, label="scatter")
plt.bar(x, y, label="bar")
plt.plot(x, y, label="plot")
plt.legend() # 

plt.title("Test Graph", size=15)                 # 제목
plt.xlabel('x')                                  # x축 범례
plt.ylabel('y')                                  # y축 범례

plt.show()
````

## 4. kor을 막대그래프로 비교하기

````python
df = pd.read_csv('./data/scores.csv')
name = df['name']
kor = df['kor']
````

````python
# plot 제목을 한글로 쓰면 깨지는 문제를 해결하기 위해 다음 코드를 먼저 실행
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = 'False'
````

````python
plt.bar(name, kor)
plt.title("국어 성적")
plt.xlabel('name')
plt.ylabel('kor_score')

plt.show()
````

위의 그림을 보면, 이름이 겹쳐서 짤린다. 다음과 같이 수정한다. 

````python
plt.bar(name, kor)
plt.title("국어 성적", size=20)
plt.xlabel('name')
plt.ylabel('kor_score')

plt.xticks(rotation=90)

plt.show()
````
