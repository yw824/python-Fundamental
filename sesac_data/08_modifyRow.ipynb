````python
import pandas as pd
df = pd.read_csv('./data/scores.csv')
df.index = range(100, 3100, 100)
df.head()
````

## 1. 인덱스 지정하여 행 추가 / 수정
``` df.loc[인덱스] = 추가 / 수정할 데이터 ```
- 인덱스가 존재하면 해당 인덱스의 데이터 수정
- 인덱스가 존재하지 않으면 데이터프레임의 끝에 행 추가

``` df.append(추가할 데이터, ignore_index=True) ```
- 추가할 새로운 데이터는 딕셔너리 형태 ( {컬럼1 : 값1, 컬럼2 : 값2, ,,, } )
- ignore_index : True이면 기존 인덱스는 무시하고, 인덱스가 새롭게 생성됨 - True를 전달해 주어야 한다. 

````python
new_value = {'name': 'Python', 'kor': 80, 'eng': 90, 'math': 100}
````

````python
df.append(new_value, ignore_index = True)
# 원래는 100씩 증가하다가, 다시 1부터 새로 인덱스 생성하고 데이터를 추가함
````

````python
df.tail() # 이 결과가 반영되지 않았다. 
````

````python
# 반영되게 하려면
df = df.append(new_value, ignore_index = True)
df
````

## 2. 인덱스 지정하여 추가/수정하기
```
df.loc[인덱스] = 추가할 데이터
df.loc[인덱스] = 수정할 데이터
```

- 인덱스가 존재하지 않으면 해당 인덱스의 데이터가 수정된다.
- 인덱스가 존재하지 않으면 데이터프레임의 끝에 데이터가 수정된다.
- 데이터의 형식 : 리스트로 전달

````python
# 인덱스 35에 추가
df.loc[35] = ['aaa', 70, 80, 90]
````

````python
# 인덱스 34에 추가
df.loc[34] = ['bbb', 80, 90, 100]
````

````python
df.tail() # index는 행번호가 아니라 행 이름이기 때문에 35가 앞에 올 수 있다.
````

````python
# 인덱스 30
df.loc[30] = ['ccc', 60, 70, 80]
df.tail() # python의 데이터가 수정되었다.
````

## 3. 행 삭제하기
``` df.drop(index = 삭제할 인덱스 리스트, inplace = True) ```
- index 리스트 내 존재하지 않는 인덱스가 있다면 에러 발생
- inplace : True일 시 삭제된 결과 적용

````python
# 30, 34, 35 삭제
df.drop([30, 34, 35], inplace = True)
````

````python
df.tail()
````

### 4.1 전체 인덱스명 변경하기
