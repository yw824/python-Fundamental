````python
import pandas as pd
df = pd.read_csv('./data/scores.csv')
df.head()
````

## 열 추가 / 수정하기

컬럼이 존재하면 수정, 존재하지 않으면 추가한다.
```
df['컬럼명'] = 추가 / 수정할 데이터
df['컬럼명'] = 컬럼 간 연산
```

![nn](./screenshot/15.PNG)

### 1.1 열 추가하기

````python
len(df)
````

````python
# 학생 번호 추가하기
df['no'] = range(1, len(df) + 1, 1)
df.head()
````

````python
# sum 추가
df['sum'] = df['kor'] + df['eng'] + df['math']
df.head()
````

### 1.2 열 수정하기

````python
# 학생 번호 수정하기(100부터 시작하여 1씩 증가하기)
df['no'] = df['no'] + 100
df.head()
````

## 2. 열 삭제하기
``` df.drop(columns = 삭제할 컬럼 리스트, inplace = True) ```
- 존재하지 않는 column 명을 사용하면 에러 발생
- inplace = True : 삭제된 결과 적용

````python
df.drop(columns = ['no', 'sum']).head() # 여기에서는 미리보기만 해주는 것, 그러나 실제 삭제는 되지 않았다.
````

````python
# 삭제된 내용 실제 데이터에 적용
df.drop(columns = ['no', 'sum'], inplace = True)
````
